## Stick-breaking problem:

Having a l length stick, break it twice, what is the PDF of the second break, what is the expectation of the second break.

Marking first break as X and second as Y, X has pdf:
$$f_X(x)=\frac{1}{l}\ \ 0 \le x \le l$$
Y has conditional pdf:
$$f_Y(y\vert X=x)=\frac{1}{x}\ \ 0 \le y \le x$$

joint pdf is:

$$f_{X,Y}(x, y) = f_Y(y\vert X=x)f_X(x) = \frac{1}{lx}$$

marginal pdf for y is:

$$f_Y(y)=\int_y^lf_{X,Y}(x, y)dx = \frac{1}{l}ln\frac{l}{y} \ \ 0\le y\le l$$

Expectation of final length is:

$$E[Y] = \int_0^lyf_Y(y)dy=\frac{l}{4}$$

To integral this, we use the fact:

$$\int udv = uv - \int vdu$$
which is the same as:
$$d(uv) = vdu + udv$$

## Bayes's rule

$$f_{X\vert Y}(x\vert y) = \frac{f_{Y\vert X}(y\vert x)f_X(x)}{f_Y(y)}$$

$$f_Y(y)=\int f_{X,Y}(x, y)dx$$

### combination of PMF and PDF

If X is discrete and Y is continuous. For example, X contains different quadratic functions with a PMF, Y is a distribution on axis, we know the conditional PDF $f_{Y\vert X}(y\vert x)$, what is $P_{X\vert Y}(x\vert y)$.

To get hybrid formula, we can use probablity of X and Y:

$$P_X(X=x)f_{Y|X}(y \ge Y \ge y+\delta\vert X=x)\delta = P_{X|Y}(X=x\vert y \ge Y \ge y+\delta)f_Y(y \ge Y \ge y+\delta)\delta$$

$$P_{X\vert Y}(x\vert y) = \frac{P_X(x)f_{Y\vert X}(y|x)}{f_Y(y)}$$

$$f_Y(y)=\Sigma p_X(x)f_{Y\vert X}(y\vert x)$$

if the observable is discrete conditional random variable, and we know the PDF of continuous random varible, then we swap above formula:

$$f_{X\vert Y}(x\vert y) = \frac{f_X(x)p_{Y\vert X}(y|x)}{p_Y(y)}$$

$$p_Y(y)=\int f_X(x)p_{Y\vert X}(y\vert x)dx$$

## Derived probability

Knowing PDF of X as an uniform distribution [0, 2], what is PDF for Y which $Y=X^3$

To solve this problem, we need CDF to help us.

$$F_Y(y) = P(Y \le y)$$

Then, we differentiate CDF $F_Y(y)$ to get PDF:

$$f_Y(y)=\frac{dF_Y(y)}{dy}(y)$$

In the example, $P(Y\le y)=P(X^3\le y)=P(X\le y^{1/3})=\frac{1}{2}y^{1/3}$, so $f_Y(y)=\frac{1}{6y^{2/3}}$

Another approach: if the probability of hitting $[x, x+\delta]$ is P, then if we have a strictly monotonic function g, which makes $y=g(x)$, then the probability of hitting $[g(x), g(x+\delta)]$ is also P:

$$\delta f_X(x)=\delta \frac{dg}{dx}(x) f_Y(y)$$

$$f_Y(y) = (\frac{dg}{dx}(g^{-1}(y)))^{-1}f_X(g^{-1}(y))$$

## Convolution

## Covariance

$$Cov(X, Y) = E[(X-E(x))(Y-E(Y))]$$

If X is far away from expectation, what about Y.

Suppose X and Y has zero expectation, $Cov(X, Y)=E[XY]$, if X is positive and Y will be also positive, then Cov is positive, vice versa.

Easy to prove:
$$Cov(X, Y) = E[XY]-E[X]E[Y]$$
$$Cov(X, X) = Var(x)$$

If X and Y are independent, then $Cov(X, Y) = 0$, converse is not true.

## Correlation coefficient

$$\rho = E[\frac{(X-E(x))}{\sigma_X}\frac{(Y-E(Y))}{\sigma_Y}]$$

$$=\frac{Cov(X, Y)}{\sigma_X\sigma_Y}$$

$$-1 \le \rho \le 1$$

If $\vert \rho \vert=1$, then, $(X-E[X]) = c(Y-E[Y])$

if independent, then $\rho=0$

## iterated expectation

$E[X|Y]$ is different from $E[X|Y=y]$, the first one is the condition is also a random varible, so the expectation itself is a random varible,  the second one is the condition is a fixed number.

$$E[E[X|Y]]=\Sigma_yE[X|Y=y]p_Y(y)$$

Using total probability theorem:
$$P(X) = \Sigma_iP(X\vert Y_i)P(Y_i)$$

$$\Sigma_yE[X|Y=y]p_Y(y) = \Sigma_y\Sigma_xP(X=x|Y=y)xp_Y(y)=\Sigma_xP(X=x)x = E[X]$$

This can help us to solve stick-breaking problem:

$$E[Y|X] = \frac{x}{2}$$

$$E[Y]=E[E[Y|X]]=\frac{1}{2}E[X]=\frac{1}{2}\frac{l}{2}=\frac{l}{4}$$

## iterated var

We know:
$$var(X|Y)=E[X^2|Y]-E^2(X|Y)$$

so:
$$E[var(X|Y)] = E[E[X^2|Y]-E^2(X|Y)] = E[E[X^2|Y]]-E[E^2(X|Y)]=E[X^2]-E[E^2(X|Y)]$$

$$var(E[X|Y]) = E[E^2[X|Y]]-E^2[E[X|Y]] = E[E^2[X|Y]] - E^2[X]$$
Add above two formula:

$$var(X) = E[var(X|Y)] + var(E[X|Y])$$
The meaning of this variance is the expectation of inter-variance in each cluster + the variance of expectation of each cluster.

## Sum of a Random number of Random varible

Suppose we hang around, and randomly shopping at some book stores, the amount of money we will spend in each book store are independent. What is the Expectation of our total budget, what is the variance for it?

N is the random variable for number of store, X is the random varible for the money we want to spend, Y is the budget we care about:

We can say:
$$E[X|N=n] = \Sigma_1^nE[X] = nE[X]$$
// $E[X]$ here is already a constant, not a random variable.
Then:
$$Y = E[X|N] = NE[X]$$

So:
$$E[Y]=E[E[X|N]] = E[NE[X]] = E[N]E[X]$$

$$var(Y) = E[var(X|N)]+var(E[X|N])$$

$$var[E[X|N]] = var(NE[X]) = E^2[X]var(N)$$

//linearity of var for independent random variable

$$E[var(X|N)] = E[Nvar(X)] = E[N]var(X)$$
So:
$$var(Y) = E^2[X]var(N)+E[N]var(X)$$
