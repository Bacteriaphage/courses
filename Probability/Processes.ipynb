## Bernoulli Process

Sequence of independent Bernoulli trial. Each trial is a binomial distribution.
So we already have lots of conclusion for Bernoulli Process, e.g. it's expectation and variance, also the property of independent for each trial and memoryless.

As a conclusion, for number of success in n stage has binomial PMF, for interarrival times has geometric PMF, Time to k arrivals has pascal PMF which is $C_{n-1}^{k-1}p^{k}(1-p)^{n-k}$

Here we need to emphasize an important point for Bernoulli process: A process is Bernoulli process only if we only have previous information, not future information, if we know any information, then it's not a Bernoulli Process.

A good example to demonstrate it is: What is the expectation of the length of the first consecutive tails in coin toss.

This question seems can be regarded as when will the first head appear(geometric distribution), because of memoryless. But we need to think about which part of the process has a geometric distribution:

The answer should be the second. Why the first one is not? Because we don't know what will happen in the future, and current result is head. if we say next toss is the start of geo distribution process, that mean we already know next toss is tail. the second is Bernoulli process because the current toss is tail, we can make sure our process is same as first head appear problem.

## Poisson Process
We can regard Poisson process as a continuous Bernuolli process, and it the stage in Bernoulli process equals to extremely small time period $\delta$. Because of the time period is so small, there are only one event can happen in this time interval, so we only have two scenarios, one is event happens, the other is event does not happen.

Poisson process also has indenpendent property that is numbers of arrivals in disjoint time intervals are independent.

We define a $\lambda$ as arrival rate which gives $\lambda \delta$ as the probability of event happens in a small time interval. In other word, arrival rate $\lambda$ is the number of arrivals in unit time.

Our number of arrivals in time $t$ problem has probability:

$$P(k, t) = lim_{n->\infty}C_n^k(\lambda \delta)^k(1-\lambda \delta)^{n-k}$$
where
$$t=n\delta\ \ p=\lambda \delta$$
then:
$$P(k, t) = lim_{n->\infty}C_n^k(\frac{\lambda t}{n})^k(1-\frac{\lambda t}{n})^{n-k}$$

$$P(k, t) = \frac{(\lambda t)^ke^{-\lambda t}}{k!}$$

Notice the expectation of number of success in time t for Poisson process is $\lambda t$ and the expectation of the time when the first success happens is $\frac{1}{\lambda}$, this two are from the defination of arrival rate.

Our Time to kth arrival problem has PDF:

$$f(t) \delta = P(k-1,t)\lambda \delta$$
$$f(t) = \frac{(\lambda t)^{k-1}}{(k-1)!}e^{-\lambda t}\lambda$$

When k = 1, this is a exponential distribution, which has the same shape as geometric distribution in Bernuolli process.

### Observation bias

the central idea is the oberservation and the expectation calculation are from different angle. for example, a survey shows the average family has 4 people, but another survey shows people are in a family with average size 6. what cause the difference? This is because of the first survey has target as each family and the second has target as people. for the second survey, it very likely to interview a people who is in a large family rather than a small one, so the bias happens.

Same reason for waiting for bus. Bus company say their bus arrival obey poisson and with arrival rate $\lambda$, so the average waiting time should be $\frac{1}{\lambda}$, but in reality, customers always ask others how long time does the bus not appear and count for his own waiting time and finally find the total waiting time is longer than what the company said. This is because customers has more possibility coming into the larger interval thana small one. If customer always come as the exactly same poisson process, which is coming after a bus goes away, then they will see the same expectation as the company said.  

## Markov Process

The concept is not hard, the probability of next move only depends on current state, which means current state is enough.

Something need to be clear, in Markov chain, there are two set of states, one is recurrent state and the other is transient state. There also another catagory of state called periodic state.
### recurrent state
This state means we can come back to the state after a period of time.
### transient state
This state means if we go out from this state, we maight not come back again.
### periodic state
This state is a recurrent state, which can be grouped into d>1 groups so that all transactions from one group lead to the next group.

### Markov steady state

A Markov process has steady state if and only if there are only one recurrent state and the recurrent state has no periodic state.

If Markov process has multiple recurrent state, then the initial state will effect the final state.
