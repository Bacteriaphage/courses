## Risk parity
Intuition: modern portfolio theory tells us we can define our return or risk(real risk is hard, use stddev of portfolio) and then we can know the expected risk or return. But risk parity strategy has another angle to look at this problem. Instead of look at a value of return or risk in the portfolio, why not balance the risk? Can we give each asset a ratio then make their risk contribution the same.

Here is a good article:
https://support.wealthfront.com/hc/en-us/articles/360000117963-How-does-Risk-Parity-work-

Take away of this article is:
1. To use risk parity strategy we must have low risk low return asset has larger sharp ratio than high risk high return asset.
2. the rate of leverage could not be so high, otherwise we cannot make money.

something not in this article:
3. the reason why this strategy can get better performance is (first plot is equity and bond, second is leverage R1 vs asset R2):


why we can draw a straight line in our efficient frontier? That is because once we calcuate the weight of asset to build up equality, we can regard this asset combination as a single asset. Then because leverage (borrow money from bank) has a const rate and 0 risk, then we can use the model in the bottom.

why the risk will go up when we have leverage? $$\sigma_p^2=\omega_1^2\sigma_1^2+\omega_b^2\sigma_b^2+2\rho_{1b}\omega_1\sigma_1\omega_b\sigma_b$$
leverage has no risk and covariance of leverage and asset is 0, so our total portfolio risk is:
$$\sigma_p = \omega_1\sigma_1$$
when leverage exist, $\omega_1+\omega_b = 1$ where $\omega_b$ can be negative, then $\omega_1$ can be greater than 1, and total risk will go up linearly.

this picture also shows why high rate leverage will fail our strategy, if we move cash higher on the y axis, our return will less than traditional 40/60 return.

Another important observation is the slope of that line in efficient frontier is the sharp ratio of current asset combination. the ratio will not change as long as the we use the same combination and how much we leverage doesn't matter.
