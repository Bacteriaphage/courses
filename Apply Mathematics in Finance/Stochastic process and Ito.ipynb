## Stochastic process

### simple random walk

Each step can go to 1 or -1, they have the same probability, then:

$$E[X_i] = 0$$
$$var(X_i) = 1$$

$$S_n = \Sigma_{i=1}^nX_i$$

$$E[S_n] = 0$$
$$var(S_n) = n$$

Because of central limit theorem, the $Z_n = \frac{S_n}{\sqrt{n}}$ is a standard normal distribution N(0, 1), and $S_n$ is N(0, n)

simple random walk is a Bernuolli process, so it also has all properties of Bernuolli process like memoryless.

Think about the random walk, the game will stop either at 100 or -50, what is probability for 100 head?
This problem is a more like a Markov process since 100 and -50 are two recurrent states, the probability can be affected by the initial state.

One way to think about it is:
for $-50<k<100$, $p(k)$ is the probability of reaching 100 at position k.
then: $p(k) = \frac{1}{2}p(k-1) + \frac{1}{2}p(k+1)$
and $p(100)=1\ \ p(-50)=0$

Calculate:


$$p(k) = \frac{1}{2}p(k-1) + \frac{1}{2}p(k+1)$$
$$p(k+1)-p(k) = p(k)-p(k-1)$$
define $S(k) = p(k)-p(k-1)$
we can easily get $S(k) = S(k-1) = ...$
suppose our upper bound is A and lower bound is 0, then:
$$p(A) =\Sigma_{k=1}^AS(k)+p(0)$$
use constant S to replace S(k):
$$1 = AS+0$$

$$S = \frac{1}{A}$$
then, for any $0<k<A$
$$p(k) = \Sigma_{n=1}^kS+p(0)$$
$$p(k) = \frac{k}{A}$$

above problem can be regard as starting from 50, stopping at 0 or 150, so we can easily get $p(0) = \frac{1}{3}$


### Martingale

A stochastic process is a martingale if:
$$X_t = E[X_{t+1}\vert F_t]$$
where
$$F_t = \{X_0, X_1, ..., X_t\}$$

For example simple random walk is a martingale. If we know the path, the expectation of next step is the current state.

### Optional stopping theorem
#### stopping time
Givng a stochastic process X, a non-negative integer random varible $\tau$ is called a stopping time if $\forall k\ge 0$, event $\tau\le k$ depends only on $X_0,...,X_k$

https://en.wikipedia.org/wiki/Optional_stopping_theorem

Suppose we have a martingale X. $\tau$ is a stopping time, furthermore, $\exists const T$ such that $\tau \le T$, then:

$$E[X_\tau] = X_0$$

go back to our 100 and -50 example. Our $E[X_\tau] = p*100+(1-p)(-50) = 0$
$p = \frac{1}{3}$

## Continuous time stochastic process

### Brownian motion
like simple random walk but has some property need to pay attention

Exists probability distribution over the set of continuous function B: $R_{\ge 0}->R$ (this means all possible function/path)
1. always start from zero $P(B(0)=0) = 1$
2. $\forall\ 0 \le s < t$
$$B(t)-B(s)->N(0, t-s)$$
3. if interval [$s_i$,$t_i$] are disjoint interval, then $B(s_i)-B(t_i)$ are independent for different i

properties:

1. crosses the time axis infinitely often
2. does not deviate too much from $t=y^2$, because of N(0, t)
3. is not differentiable, but can be analysed by Ito's calculus.

extra property:
assume $B(t)$ is a brownian motion and $M(t)$ is an event of max{$B(t)$}, for all $t>0$ and $a\ge 0$
$$P(M(t)>a) = 2P(B(t)>a)$$

proof.

make $\tau_a = min_t\{B(t)=a\}$
$$P(M(t)>a)=P(\tau_a<t)$$
$$=P(B(t)-B(\tau_a)>0\vert \tau_a<t)+P(B(t)-B(\tau_a)<0\vert \tau_a<t)$$
because of symmetric
$$=2P(B(t)-B(\tau_a)>0\vert \tau_a<t)$$
$$=2P(B(t)-a>0\vert \tau_a<t)$$
because of continuity of brownian motion
$$=2P(B(t)>a)$$

### quadratic variance
For a Brownian motion $B(t)$:
$$lim_{n->\infty}\Sigma_{i=1}^n(B(\frac{i}{n}T)-B(\frac{i-1}{n}T))^2 = T$$
This can be rewrite as:
$$(dB)^2=dt$$
This is the very important property we use to apply Ito's lemma.

Any differentiable function does not have this property, because:
$$\Sigma(f(t_{i+1}-f(t_i))^2 \le \Sigma(f'(S_i)(t_{i+1}-t_i))^2$$
mean value theorem
$$=(max_{0\le S \le T}f'(s)^2)*\Sigma(t_{i+1}-t_i)^2$$
$$=\ const*\frac{T_2}{n}->0$$

proof of quadratic variance
make $t_i = \frac{i}{n}T$
$$\Sigma_{i=1}^n(B(\frac{i}{n}T)-B(\frac{i-1}{n}T))^2 =\Sigma_{i=1}^n(B(i)-B(i-1))^2$$
make $X_i = B(i+1)-B(i)$, then X is a normal distribution r.v. $N(0, \frac{T}{n})$.
$$E[X_i^2] = var(X_i) = \frac{T}{n}$$

$$\Sigma_{i=1}^n(B(i)-B(i-1))^2 = \Sigma_{i=0}^{n-1} X_i^2$$
$$=n(\frac{\Sigma_{i=0}^{n-1} X_i^2}{n})$$
weak law of large number
$$=n\frac{T}{n} = T$$


## Ito's calculus
Let $f(x)$ as well defined function, differentiable, continuous etc. $B_t$ is a brownian motion. and we need to calculate $df(b_t)$

giving taylor expansion:
$$f(x+a)=f(x)+f'(x)a+\frac{f''(x)}{2!}a^2+\frac{f'''(x)}{3!}a^3...$$
$$f(x+a)-f(x)=f'(x)a+\frac{f''(x)}{2!}a^2+\frac{f'''(x)}{3!}a^3...$$
$$f(B_{t+1})-f(B_{t}) = f'(B_t)dB_t+\frac{f''(B_t)}{2}(dB_t)^2+\frac{f'''(B_t)}{3!}(dB_t)^3$$
because of quadratic variance
$$df = f'(B_t)dB_t+\frac{f''(B_t)}{2}dt$$

This is prototype of $Ito's\ Lemma$

In practice, $f$ can be a well defined price funtion like call option and $B$ can be an underlying asset like stock price.

BTW, if $B$ is not a brownian motion and just a differentiable function, then $(dB)^2$ will go to 0 if $dB$ is small.

### simple Ito's lamma
$$f(t+dt, S+dS) = f(t, S)+\frac{\partial f}{\partial t}dt+\frac{\partial f}{\partial S}dS+\frac{1}{2}(\frac{\partial^2 f}{\partial t^2}(dt)^2+2\frac{\partial^2 f}{\partial t\partial S}dtdS+\frac{\partial^2 f}{\partial S^2}(dS)^2)$$
apply quadratic variance:
$$df = f(t+dt, S+dS)-f(t, S)=\frac{\partial f}{\partial t}dt+\frac{\partial f}{\partial S}dS+\frac{1}{2}\frac{\partial^2 f}{\partial S^2}(dS)^2$$
$$=\frac{\partial f}{\partial t}dt+\frac{\partial f}{\partial S}dS+\frac{1}{2}\frac{\partial^2 f}{\partial S^2}dt$$
### full Ito's lemma
If $S$ is a Brownian motion with drift:
$$dS = \mu dt+\sigma dW$$
plug it into above equition, we get:
$$df(t, S_t) = (\frac{\partial f}{\partial t}+\mu \frac{\partial f}{\partial S}+\frac{1}{2}\sigma^2\frac{\partial^2 f}{\partial S^2})dt+\frac{\partial f}{\partial S}dW$$

## Ito integral
remember how we calcucate integral for normal function, we split whole interval to extremely small one and use value of the interval(min, max, left, right) to sum up, and the final result will converge to the real result. But for Brownian motion, we can only use left value of the interval, that is because we know no information about the future, we only know current state which is the left value.

### concept of adapted

$\Delta(t)$ is adapted to $X_t$ for all $t \ge 0$, $\Delta(t)$ only depends on $X_0$~$X_t$, this fit Ito calculus very well.

suppose $B(t)$ is a brownian motion at time t:
$$B(t):\ N(0, t)$$
if $X(t) = \sigma B(t)$, then $X(t)$ also has normal distribution $N(0, \sigma^2t)$, which can be rewrite as:
$$X(t)=\int \sigma dB_t$$
This can be think as the sum of the normal distribution is also a normal distribution.

Theorem:
if $\Delta(t)$ is a process depends only on time t, then:
$$X(t)=\int \Delta(t)dB(t)$$
has normal distribution at all time.

### Ito isometry
if $\Delta(t)$ adapted to $B_t$
$$E[(\int_0^t\Delta(s)dB_s)^2]=E[\int_0^t\Delta^2(t)ds]$$

if $\Delta(s) =1$, then:
$$E[(B_t)^2]=t$$
which is quadratic variance

### martingale in Ito integral
if $g(t, B_t)$ is adapted to $B_t$ then $\int g(t, B_t)dB_t$ is a martingale as long as g is "reasonable", which $\int\int g^2(t, B_t)dtdB_t<\infty$ 

for an asset $X_t$:
$dX_t = \mu(t, B_t)dt + \sigma (t, B_t)dB_t$

if $\mu=0$, then $X_t$ is a martingal.

## Change of measure
### Radon-Nikodym derivative
$P$ and $\tilde P$ are two probability space apply to a set of event $\Omega$, for any subset $\omega$, if $P(\omega)>0$ and $\tilde P(\omega)>0$, then these two probability space are equivalent, which means we can find a transform funtion $Z(\omega)$ to get:
$$P(\omega)=Z(\omega)\tilde P(\omega)$$
We can imagine if one of $\tilde P(\omega)$ is 0, the $Z$ will not exist.

### Girsonov's theorem
if we have to Brownian motion end at $T$, one $P$ has drift $\mu$, another $\tilde p$ has no drift, they are also in the equivalent probability space, which mean we can find a Radon-Nikodym derivative $Z$ to link them together.
And we have:
$$Z(\omega) = \frac{d\tilde P}{dP}(\omega)=e^{-\mu\omega(T)-\mu^2T/2}$$

Application for this:

If we have a portfolio $V_t$ which under stock price measure with drift, we can use Girsonov to change measure to a driftless martingale measure, to compute the expectation of price, we have:
$$\tilde E[V_t] = E[Z(t)V_t(t)]$$
