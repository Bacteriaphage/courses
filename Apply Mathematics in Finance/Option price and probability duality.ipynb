## Derivatives
### Call option

payoff: $max(S_T-K, 0)$, price: $C_K(t, T)$


The price is zero when S is zero, when S is infinity, the price is the discount of S-K.

### digital option
payoff: $I\{S_T>K\}$, price: $D_K(t, T)$

Consider a portfolio: $\lambda$ calls with strike $K$ and $-\lambda$ calls with strike $K+\frac{1}{\lambda}$

Price function for this portfolio is: $\lambda C_K(t, T)-\lambda C_{K+\frac{1}{\lambda}}(t, T)$

This price function can be rewrite as:
$$D_K(t, T) = lim_{\epsilon->0}\frac{C_K(t, T)-C_{K+\epsilon}(t, T)}{\epsilon} = -\frac{\partial C_K(t, T)}{\partial K}$$
where $\epsilon=\frac{1}{\lambda}$

To comprehend this result, at time T, we want a digital payoff, and at time T, the slope of call option goes from 0 to $\frac{dC_K(T, T)}{dK}$ which is a digital jump. Since we can use this slope to construct digital option, we can also use the price of this slope to price digital option.



## Fundemental theorem of asset price
1. A discrete market, on discrete probability space $(\Omega, F, P)$, is arbitrage-free if and only if there exists at least one risk neutral probability measure that is equivalent to the original probability measure P.
2. An arbitrage free market (S, B) consisting of a collection of stock S and a risk free bond B is complete if and only if there exists a unique risk-neutral measure that is equivalent to P and has numeraire B.

By the FTAP, price at t is the expected payout at T, suitably discounted. for example, we we know the payoff function of call options and we also know the current stock price, we can calculate expected value of payoff and apply a discount factor to it, then we get current price. As we can see, this is based on arbitrage-free.

rewrite digital price function, $Z(t, T)$ is the zero coupon bond price and it is used as a discount factor:
$$D_K(t, T) = Z(t, T)E[I\{S_T>K\}\vert S_t]$$
strictly $\frac{D_K(t, T)}{Z(t, T)}$ is martingale.
take one more step:
$$D_K(t, T) = Z(t, T)P(S_T>K\vert S_t)$$
then:
$$P(S_T<K\vert S_t) = 1+\frac{\partial C_K(t, T)}{\partial K}\frac{1}{Z(t, T)}$$
the we can get density function:
$$f_{S_T}(x)=\frac{1}{Z(t, T)}\frac{\partial^2C}{\partial K^2}\vert_x$$
So, now given a call option price and strike, we can know the probability density function of underlying asset at time T in the condition of current price is $S_t$.

## Call butterfly
consider another portfolio:

$\lambda$ calls with strike $K-\frac{1}{\lambda}$, $-2\lambda$ calls with strike $K$ and $\lambda$ calls with strike $K+\frac{1}{\lambda}$

we call this call butterfly, the price of this contract:
$$B_{k, \lambda}(t, T) = \lambda (C_{K-\frac{1}{\lambda}}(t, T)-C_K(t, T))-\lambda (C_K(t, T)- C_{K+\frac{1}{\lambda}}(t, T))$$
and
$$\lambda B_{k, \lambda} \approx \frac{\partial^2C}{\partial K^2}$$
then we can rewrite it as:
$$B_{k, \lambda} \approx \frac{1}{\lambda}f_{S_T}(x)\vert_K$$

This butterfly has a payoff function as a triangle around strike K, by estimate the stock price at time T, we can find whether the price of butterfly is good or not, then we can buy or sell.

## Replicate portfolio of any derivatives
For a derivative with payoff $D(T,T)=g(S_T)$ (D is for derivative, not only for digital), price function $D(t, T)$ at $t \le T$ has:

$$\frac{D(t, T)}{Z(t, T)}=E[\frac{D(T,T)}{z(T, T)}\vert S_t]$$
(martingale)

$$D(t, T)=Z(t, T)\int g(x)f_{S_T\vert S_t}(x)dx$$
So, if we know the density then we know the derivative price.

Think about the relationship between call option and density function. We somehow understand call option price span all derivative prices.

If we have a derivative which has piecewise linear payoff function, then we can construct a portfolio of call options to replicate that payoff function at time T. According to FTAP, if two assets have the same payoff at time T, then they should have the same price with discount factor at time t. So our portfolio of call options should has the price funtion as our new derivative. (A good example is digital and butterfly)

To be noticed, there is no reason why the replication portfolio and the new derivative will have same price all the time, but in theory, at time T, if we build portfolio in above way, we can get the same result.

If our payoff function is a smooth funtion, we have also use the same idea but instead of $\Sigma$, we need $\int$.

### From taylor's aspect
$$g(x) = g(0)+xg'(0)+\int_0^\infty (x-c)^+g''(c)dc$$
replace dummy varible with k and x with $S_T$
$$g(S_T) = g(0)+S_Tg'(0)+\int_0^\infty (S_T-K)^+g''(K)dK$$
take discounted expected value:
$$D(t,T)=Z(t, T)g(0)+S_tg'(0)+\int_0^\infty C_K(t, T)g''(K)dK$$
So a derivative can be replicated by $g(0)$ zero coupon bond, $g'(0)$ stock, and a portfolio of call options.
